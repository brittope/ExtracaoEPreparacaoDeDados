```python
import easyocr
import cv2
import matplotlib.pyplot as plt
from docling.document_converter import DocumentConverter
```

```python
imagem_path = '../DataFrames/carro.jpg'

reader_en = easyocr.Reader(['pt'])

# Fazer leitura
results_en = reader_en.readtext(imagem_path, decoder="beamsearch", mag_ratio=5, allowlist='ABCDEFGHIJKLMNOPQRSTUVWYZ1234567890')


# 3. Mostrar a imagem
image = cv2.imread(imagem_path)
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

print("\n--- Resultados usando idioma português ['pt'] ---")
for detection in results_en:
    print(f"Texto detectado: {detection[1]} - Confiança: {detection[2]:.2f}")
```
