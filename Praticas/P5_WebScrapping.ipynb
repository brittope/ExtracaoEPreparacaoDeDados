```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Configura o Chrome em modo headless
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Inicializa o driver do Chrome
service = Service()  # se precisar pode passar o caminho do chromedriver: Service("caminho/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

# Abre a página
driver.get("https://ge.globo.com/futebol/brasileirao-serie-a/")

# Aguarda a página carregar (pode ajustar o tempo se necessário)
time.sleep(20)

# Pega o HTML da página
html_content = driver.page_source

# Fecha o navegador
driver.quit()

# Extrai todas as tabelas com pandas
tabelas = pd.read_html(html_content)

print(f"{len(tabelas)} tabelas encontradas.")
```

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio
from playwright.async_api import async_playwright
import pandas as pd
import time

async def obter_tabelas():
    async with async_playwright() as p:
        browser = await p.firefox.launch(headless=True)
        page = await browser.new_page()
        await page.goto('https://www.fundsexplorer.com.br/ranking')
        time.sleep(20)

        # Pega o HTML completo da página
        html_content = await page.content()
        await browser.close()

    # Usa pandas.read_html para extrair todas as tabelas do HTML
    tabelas = pd.read_html(html_content)

    print(f"{len(tabelas)} tabelas encontradas.")
    return html_content, tabelas

# Executa no loop existente
html_content, tabelas = await obter_tabelas()
```

```python
import requests
import json
import pandas as pd

url = "https://www.fundsexplorer.com.br/wp-json/funds/v1/get-ranking"

headers = {
    "accept": "application/json, text/plain, */*",
    "referer": "https://www.fundsexplorer.com.br/ranking",
    "user-agent": "Mozilla/5.0",
    "x-funds-nonce": "61495f60b533cc40ad822e054998a3190ea9bca0d94791a1da",
}

response = requests.get(url, headers=headers)

data = response.json()

if isinstance(data, str):
    data = json.loads(data)

if isinstance(data, list) and isinstance(data[0], str):
    data = [json.loads(item) for item in data]

df = pd.DataFrame(data)

df
```
